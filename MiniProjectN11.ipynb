```python
#import the necessary libraries
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import time
import re
import datetime
```

Scraping the TOP 20 Most Popular games on www.Steam.com

```python

# List of URLs to scrape
urls = [
    'https://steamcommunity.com/app/2005010/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/730/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/1085660/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/1304930/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/1811260/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/1268750/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/578080/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/1172470/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/2005010/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/1517290/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/255710/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/594570/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/289070/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/570/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/990080/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/1604030/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/359550/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/359550/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/582660/reviews/?browsefilter=toprated&snr=1_5_100010_',
    'https://steamcommunity.com/app/2167580/reviews/?browsefilter=toprated&snr=1_5_100010_'
]

# Defining the arrays (attributes)
reccs = []
hours = []
Dates = []
helpfuls = []
funnys = []
numcomments = []
GameNames = []
#defining the dataframe header
DataFrameHeaders = ['Recommended', 'hours', 'Date', 'numFoundHelpful', 'numFoundFunny', 'NumOfComments', 'GameName']
df = pd.DataFrame(columns=DataFrameHeaders)
#loop to go through every array one by one 
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'lxml')

    # Scraping the wanted data (each attribute by its class name in the url's html code)
    lists = soup.find_all('div', 'apphub_UserReviewCardContent')
    comm = soup.find_all('div', class_='apphub_CardCommentCount alignNews')

    for x in range(len(lists)):
        #Scrape the name of the game 
        GameName = soup.find('div', {'class': 'apphub_AppName ellipsis'}).string
        #Scrape 'Recommended' or 'Not Recommended'
        recc = lists[x].find('div', {'class': 'title'}).string
        #Scrape the gamer's playing hours 
        hour = lists[x].find('div', {'class': 'hours'}).string
        #Scrape the date of the comment
        Date = lists[x].find('div', {'class': 'date_posted'}).string
        #Scrape whether or not the veiwrs found the comment helpful or funny
        found = lists[x].find('div', {'class': 'found_helpful'})
        helpful = found.find('br').previous_sibling.string
        funny = found.find('br').next_sibling.string
        #Scrape the number of people who commented on the review
        numcomment = comm[x].text


        # Clean the data
        #taking only the number from the hour string + converting it to int
        cleanhour = int(float(re.sub(r"[^\d.]", "", hour)))
        #taking only the number from the helpful string + converting it to int
        cleanhelpful = int("".join(re.findall(r"\d+", helpful)))
        #taking only the number from the funny string + converting it to int
        cleanfunny = int("".join(re.findall(r"\d+", funny)))
        #Replacing 'Reccommended' to 1 and 'Not Recommended' to 0
        recc = 1 if recc == 'Recommended' else 0

        # Add the extracted values to their respective arrays
        reccs.append(recc)
        hours.append(cleanhour)
        Dates.append(Date)
        helpfuls.append(cleanhelpful)
        funnys.append(cleanfunny)
        numcomments.append(numcomment)
        GameNames.append(GameName)

# Create DataFrame from the extracted data
data = {
    'Recommended': reccs,
    'hours': hours,
    'Date': Dates,
    'numFoundHelpful': helpfuls,
    'numFoundFunny': funnys,
    'NumOfComments': numcomments,
    'GameName': GameNames
}
df = pd.DataFrame(data)

# Save DataFrame to a CSV file
df.to_csv('gaming.csv', index=False)

# Print the DataFrame
print(df)
```
